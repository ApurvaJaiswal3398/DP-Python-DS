```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
df =  sns.load_dataset('iris')
df
```

```python
# bar plot for averge sepal length (by default)
_ = sns.barplot(df, x='species', y='sepal_length')  # _ is the garbage variable
```

```python
# bar plot for maximum sepal length
_ = sns.barplot(df, x='species', y='sepal_length', estimator='max')
```

```python
_ = sns.barplot(df, x='species', y='sepal_length', estimator='min')
```

```python
_ = sns.barplot(df, x='species', y='sepal_width')
```

```python
_ = sns.barplot(df, x='species', y='petal_length')
```

```python
_ = sns.barplot(df, x='species', y='petal_width')
```

```python
# creating subplots in rows and columns structure
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=[10, 10], nrows=2, ncols=2)
_ = sns.barplot(df, x='species', y='sepal_length', ax=ax[0,0])
_ = sns.barplot(df, x='species', y='sepal_width', ax=ax[0,1])
_ = sns.barplot(df, x='species', y='petal_length', ax=ax[1,0])
_ = sns.barplot(df, x='species', y='petal_width', ax=ax[1,1])
```

```python
# creating subplots as columns
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=[20, 5], ncols=4)
_ = sns.barplot(df, x='species', y='sepal_length', ax=ax[0])
_ = sns.barplot(df, x='species', y='sepal_width', ax=ax[1])
_ = sns.barplot(df, x='species', y='petal_length', ax=ax[2])
_ = sns.barplot(df, x='species', y='petal_width', ax=ax[3])
```

```python
# creating subplots as rows
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=[5, 20], nrows=4)
_ = sns.barplot(df, x='species', y='sepal_length', ax=ax[0])
_ = sns.barplot(df, x='species', y='sepal_width', ax=ax[1])
_ = sns.barplot(df, x='species', y='petal_length', ax=ax[2])
_ = sns.barplot(df, x='species', y='petal_width', ax=ax[3])
```
