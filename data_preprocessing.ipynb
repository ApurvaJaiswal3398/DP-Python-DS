```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('watches.csv')     # Load Data
df.info()       # Preprocessing of Data
```

```python
df
```

```python
def clean_price_column(price):
    if isinstance(price, str):
        price = price.replace(',', '')
        return float(price)
    return price
```

```python
def clean_oprice_column(oprice):
    if isinstance(oprice,str):
        if('₹' in oprice):
            oprice=oprice.replace(',','').replace('₹','')
        else:
            oprice=np.nan
        return float(oprice)
    return float(oprice)
```

```python
clean_oprice_column('Not in stock')
```

```python
df['oprice'].apply(clean_oprice_column)
```

```python
df['clean_oprice']=df['oprice'].apply(clean_oprice_column)
```

```python
df
```

```python
def clean_rating_column(rating):
    if isinstance(rating,str):
        if('.' in rating) and (len(rating)==3):
            return float(rating)
        else:
            return np.nan
    return float(rating)
```

```python
clean_rating_column('hello')
```

```python
df['clean_rating']=df['rating'].apply(clean_rating_column)
```
