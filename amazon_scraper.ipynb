```python
#import web mining library
from dputils.scrape import Scraper, Tag
```

```python
page = 1
query = 'Laptop'
#register the webpage for scraping
webpage = Scraper(f'https://www.amazon.com/s?k={query}&page={page}')
#Display data in <title> tag
webpage.soup.title
```

```python
#Display the html code of the page
webpage.soup
```

```python
# Select the target tag to scrape the data
target = Tag('div', cls='s-main-slot s-result-list s-search-results sg-row')
# By default 'div' tag is selected
items = Tag(cls='s-result-item')
title = Tag('h2')
price = Tag('span', cls='a-price-whole')
rating = Tag('span', cls='a-size-base')
#retrieving the needed tags
scrapedata = webpage.get_all(target, items, name=title, price=price, rating=rating)
for i in scrapedata:
    if i['name'] and i['price'] and i['rating']:
        print('\n' + i['name'])
        print(f"Price : ${i['price'].replace('.', '')}")
        print(f"Rating : {i['rating']} / 5.0")
    # print(i)
```

```python
for i in webpage.get_all(target, items, name=title, price=price, rating=rating):
    if i['name'] and i['price'] and i['rating'] and int(i['price'].replace('.', '').replace(',',''))>=800:
        print('\n' + i['name'])
        print(f"Price : ${i['price'].replace('.', '')}")
        print(f"Rating : {i['rating']} / 5.0")
```

```python
sd = [i for i in scrapedata if i['name'] and i['price'] and i['rating'] and int(i['price'].replace('.', '').replace(',',''))>=200]
for i in sorted(sd, key=lambda i: ( i['rating'], int(i['price'].replace('.', '').replace(',','')))):
    print('\n' + i['name'])
    print(f"Price : ${i['price'].replace('.', '')}")
    print(f"Rating : {i['rating']} / 5.0")
```
